###  "Estamos utilziando Box-plot, que possui a seguinte configuração: temos uma mediana, em que dividimos os dados em 50%, para a direita e esquerda, igualmente. Teremos o Q1, que se refere ao primeiro quartil e parte em 25% e 75%, já o Q3 parte os dados em 75% e 25%. A diferente entre Q3 e Q1 gera o intervalo interquartirico, isto é, as estatísticas que geraremos para realizar o corte de outliers."

```python
valor = dados['Valor']
```

```python
#Cálculo dos quartis
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ  
limite_superior = Q3 + 1.5 * IIQ  
```

```python
selecao = (valor >= limite_inferior) & (valor <= limite_superior)  #Seleção com valores dentro dos intervalos
dados_new = dados[selecao]  #Novo dataframe com os novos valores da selecao
```

```python
#Grupamento somente da variável Valor, por Tipo de imóvel
grupo_tipo = dados.groupby('Tipo')['Valor']
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ  
limite_superior = Q3 + 1.5 * IIQ 
```

```python
Q1
```

```python
Q3
```

```python
IIQ
```

```python
#Loop nas chaves 
for tipo in grupo_tipo.groups.keys():
    print(tipo)
```

```python
#Criar box-plot para cada tipo de imóvel, com os valores nos intervalos calculados
dados_new = pd.DataFrame()  #Zerar o dataframe dados_new
for tipo in grupo_tipo.groups.keys():   #Loop nas chaves do grupamento (Tipos)
    eh_tipo = dados['Tipo'] == tipo     #Variável que representa cada tipo de imóvel(identificado no for)
    #Variável correspondente aos valores que estão entre o limite inferior e superior
    eh_dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])  
    selecao = eh_tipo & eh_dentro_limite   #Variável correspondente à junção dos Tipos(imóveis) e Valores(intervalos)
    dados_selecao = dados[selecao]         #Novo dataframe com a variável selecao 
    
    #Dataframe que une o dados_new(zerado) e o dados_selecao, gerando um dataframe com a sequencia de todos os imóveis
    dados_new = pd.concat([dados_new, dados_selecao])  
```
