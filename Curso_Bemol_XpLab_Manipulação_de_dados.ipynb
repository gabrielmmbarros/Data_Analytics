**Função que une dois dataframes, identificando colunas em comum entre eles e trazendo esses valores em uma só coluna.**

pd.merge(**dataframe1**, **dataframe2**, on = **lista**, how = string específica)

**Parâmetros:**
1. Tabela da esquerda
2. Tabela da direita
3. lista = ['nome das colunas em comum']
4. how = método da correlação: 'left', 'right','inner(ponto em comum), 'outer'(tudo)


```
#Variável que recebe a união das duas primeiras tabelas, priorizando a df_materiais
df_relatorio = pd.merge(df_materiais, df_pedidos, on = ['material'], how = 'left')
#Os valores que existem em apenas um dos dois dataframes(tabelas) são colocados, na nova tabela que foi gerada, como NaN
```

```
df_relatorio
```

```
df_relatorio = pd.merge(df_relatorio, df_vendas, on = ['material'], how = 'left')
```

## Tratamento pós Correlação

```
df_relatorio.head()
```

```
#Função que trata os valores NaN da variável definida. Pode ser aplicado à variável inteira ou, no caso de dataframes, à colunas específicas
df_relatorio = df_relatorio.fillna(0)
df_relatorio
```

```
#Transformar os valores para inteiro. Caso haja algum valor NaN, a função não funciona, primeiro é necessário tratar
df_relatorio[['qtd_pedido', 'qtd_estoque', 'qtd_vendas']] = df_relatorio[['qtd_pedido', 'qtd_estoque', 'qtd_vendas']].astype(int)
```

```
df_relatorio
```

## Operações entre colunas

```
#Cria uma nova coluna com os resultados da soma entre as outras duas
df_relatorio['qtd_loja'] = df_relatorio['qtd_pedido'] + df_relatorio['qtd_estoque']
df_relatorio
```

```
#Cria uma cópia do relatorio, mas removendo as colunas de pedido e estoque, e reorganizando a ordem das colunas
df_relatorio = df_relatorio[['material', 'qtd_loja', 'qtd_vendas']].copy()
df_relatorio
```

**Função que percorre as linhas do dataframe e faz operações selecionando a coluna**

dataframe.apply(lambda row: funcao(row['coluna'], axis = '**columns**' ou **1**

```
def calcularPorcentagem(loja,vendas):
  if(vendas != 0):
    return round(loja / vendas, 2)    #Formata o valor para 2 casas decimais    
  elif(loja != 0):
    return 1.2    #Se um valor da coluna qtd_loja for diferente de 0, automaticamente já é 20%
  else:
    return 0    #Se algum dos dois valores for 0, o retorno da operação já será 0
```

```
df_relatorio['porcentagem'] = df_relatorio.apply(lambda row: calcularPorcentagem(row['qtd_loja'], row['qtd_vendas']), axis = 'columns')
```

```
df_relatorio
```

```
#Se a porcentagem for maior que 1.2(20%) uma mensagem de alerta deve ser gerada
def definirStatus(porcentagem):
  if(porcentagem >= 1.2):
    return 'alerta'
  else:
    return ''
```

```
df_relatorio['status'] = df_relatorio.apply(lambda row: definirStatus(row['porcentagem']), axis = 1)
```

```
df_relatorio
```

# Ordenar Colunas

```
#Função que ordena uma ou mais colunas de um dataframe, pelo seu valor, de forma crescente(True) ou decrescente(False)
df_relatorio.sort_values( ['qtd_loja'], ascending=[True])
```

```
df_relatorio = df_relatorio.sort_values( ['qtd_loja'], ascending=[False])
df_relatorio.head()
```

```
#Filtrando colunas
df_relatorio_reduzido = df_relatorio[(df_relatorio['status'] != 'alerta') & (df_relatorio['qtd_loja'] > 0) & (df_relatorio['qtd_vendas'] > 0)]
```

# Exportando Dados

```
# CSV ou TXT
df_relatorio.to_csv('df_relatorio_final.csv', sep = ';')
```

```
# XLSX
writer = pd.ExcelWriter('Relatorio Final.xlsx')   #Nome do arquivo

#sheet name: nome da aba(um arquivo excel pode conter várias abas)
#index = True ou False - decide se a coluna de índice que o dataframe gera, virá ou não no arquivo
df_relatorio.to_excel(writer, sheet_name = 'relatorio', index = False)   
df_relatorio_reduzido.to_excel(writer, sheet_name='reduzido', index = False)   #Cria um arquivo excel com duas abas

writer.save()   #Gerar o arquivo
```
